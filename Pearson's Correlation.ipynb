# Pearson's correlation (Filter method)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
import seaborn as sns
import matplotlib.pyplot as plt

# loading Boston Housing dataset as dictionary.
boston_dataset = load_boston()
```

```python
df = pd.DataFrame(data=boston_dataset.data, columns=boston_dataset.feature_names)
df["MEDV"] = boston_dataset.target

x = df.drop("MEDV", axis=1)
y = df["MEDV"]

x_train, x_test, y_train, y_test = \
    train_test_split(x, y, test_size=0.2, random_state=0)

df
```

**Creating a Correlation Matrix using corr() function from pandas.
It used Pearson's correlation formula to compute correlation coefficient.**

```python
correlation_matrix = x_train.corr()

correlation_matrix
```

**Creating a Correlation Matrix Heatmap.**

```python
plt.figure(figsize=(10, 12))
sns.heatmap(correlation_matrix, linewidths=1, cmap="Blues",annot=True)
plt.show()
```

**With the following function we can get list of features which 
are highly correlated with each other.
It will remove the first feature that is correlated with 
any other feature.
It takes the dataset & threshold value for correlation coefficient.
If 2 features have correlation coefficient above the threshold 
then one of them is added to the set.**

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
correlated_features = correlation(x_train,0.7)

print(len(correlated_features))
correlated_features
```

**Dropping the correlated features from the train/test dataset.**

```python
x_train.drop(correlated_features,axis=1)
x_test.drop(correlated_features,axis=1)

x_train
```
