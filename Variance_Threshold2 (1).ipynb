# Variance Threshold (Filter method)
### Dataset used : https://www.kaggle.com/c/santander-customer-satisfaction/data?select=train.csv

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold

# It has 76k rows , 371 columns , hence taking only 10k rows.
df = pd.read_csv("santander_dataset.csv", nrows=10000)

x = df.drop("TARGET", axis="columns")
y = df["TARGET"]

train_x, test_x, train_y, test_y = \
    train_test_split(x, y, test_size=0.2, random_state=0)

print(train_x.shape , test_x.shape)
```

```python
train_x
```

```python
var_threshold = VarianceThreshold(0)

var_threshold.fit(train_x)

optimal_columns = train_x.columns[var_threshold.get_support()]
constant_columns = []
```

```python
optimal_columns
```

```python
for column in train_x.columns:
    if column not in optimal_columns:
        constant_columns.append(column)


df_optimal = df.drop(constant_columns, axis="columns")
df_constants = df.drop(optimal_columns, axis="columns")

print("Total Columns containing constants : {}".format(len(constant_columns)))
```
